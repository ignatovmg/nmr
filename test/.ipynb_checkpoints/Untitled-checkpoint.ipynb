```python
% matplotlib inline

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
comp = pd.read_csv("/home/mikhail/Projects/nmr/sandbox/computed_matrix", sep = '\t', header=None)
empty = pd.DataFrame([[0.0]*comp.shape[0]]*comp.shape[1])
for i in range(comp.shape[0]):
    comp.iloc[i, i] = 0.0
```

```python
exp_stacked = pd.read_csv("/home/mikhail/Projects/nmr/sandbox/matrix/00000", sep = '\t', header=None)
exp = (empty + exp_stacked.pivot(0, 1, 2)).fillna(0.0)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title("Experiment")
sns.heatmap(exp, linewidths=0.01)
plt.subplot(1, 2, 2)
sns.heatmap(-comp, linewidths=0.01)
plt.title("Computed (good complex CIL_04844)")
```
